```python
# Import pandas, matplotlib.pyplot, and seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Local location of the data
# local_data = 'D:\\Data_Science\\DroughtProject\\Data\\' # Location on Windows 
local_data = '/Users/tanishalohchab/Downloads/drought_prediction-main/dataset/' # Location on Linux

# The data is split into four csv files.
# Load all four files
soil_set = pd.read_csv(local_data + 'soil_data.csv')
test_set = pd.read_csv(local_data + 'test_timeseries.csv')
train_set = pd.read_csv(local_data + 'train_timeseries.csv')
validation_set = pd.read_csv(local_data + 'validation_timeseries.csv')
```

```python
validation_set['date'] = pd.to_datetime(validation_set['date'])
```

```python
# print(validation_set.date.describe(datetime_is_numeric=True))
# Ensure 'date' column is in datetime format
validation_set['date'] = pd.to_datetime(validation_set['date'], errors='coerce')

# Compute numeric statistics manually
date_min = validation_set['date'].min()
date_max = validation_set['date'].max()
date_range = date_max - date_min
date_mean = validation_set['date'].mean()
date_median = validation_set['date'].median()

# Display statistics
print(f"Min: {date_min}")
print(f"Max: {date_max}")
print(f"Range: {date_range}")
print(f"Mean: {date_mean}")
print(f"Median: {date_median}")
```

```python
# test_set = pd.read_csv(local_data + 'test_timeseries.csv')
# test_set['date'] = pd.to_datetime(test_set['date'])
# print(test_set.date.describe(datetime_is_numeric=True))
# Read CSV and convert 'date' column to datetime
test_set = pd.read_csv(local_data + 'test_timeseries.csv')
test_set['date'] = pd.to_datetime(test_set['date'], errors='coerce')

# Compute summary statistics manually
date_min = test_set['date'].min()
date_max = test_set['date'].max()
date_range = date_max - date_min
date_mean = test_set['date'].mean()
date_median = test_set['date'].median()

# Print results
print(f"Min: {date_min}")
print(f"Max: {date_max}")
print(f"Range: {date_range}")
print(f"Mean: {date_mean}")
print(f"Median: {date_median}")
```

```python
# Read CSV and convert 'date' column to datetime
train_set = pd.read_csv(local_data + 'train_timeseries.csv')
train_set['date'] = pd.to_datetime(train_set['date'], errors='coerce')

# Compute statistics manually
print("Datetime Column Statistics:")
print(f"Min: {train_set['date'].min()}")
print(f"Max: {train_set['date'].max()}")
print(f"Range: {train_set['date'].max() - train_set['date'].min()}")
print(f"Mean: {train_set['date'].mean()}")
print(f"Median: {train_set['date'].median()}")
```
